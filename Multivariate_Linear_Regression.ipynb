```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/ML Lab Datasets/50_Startups.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.histplot(df['Profit'],bins=30,kde=True, stat="density")
```

```python
numerical_df = df.select_dtypes(include=['number'])
```

```python
corr = numerical_df.corr().round(1)
```

```python
sns.heatmap(data=corr,annot=True)
```

```python
plt.figure(figsize=(20,5))
features=['R&D Spend','Profit']
target=df['Profit']
for i, col in enumerate(features):
    plt.subplot(1, len(features), i + 1)
    x = df[col]
    y = target
    plt.scatter(x, y, marker="o")
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Profit")
```
