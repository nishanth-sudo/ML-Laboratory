```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/ML Lab Datasets/Student_Performance.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.histplot(df['Performance Index'],bins=30,kde=True, stat="density")
```

```python
for column in df.select_dtypes(include=['object']).columns:
    if df[column].isin(['Yes', 'No']).all():  # Check if column contains only 'Yes' and 'No'
        df[column] = df[column].map({'Yes': 1, 'No': 0})
```

```python
corr=df.corr().round(1)
```

```python
sns.heatmap(data=corr,annot=True)
```

```python
plt.figure(figsize=(20,5))
features=['Hours Studied','Performance Index']
target=df['Hours Studied']
for i, col in enumerate(features):
    plt.subplot(1, len(features), i + 1)
    x = df[col]
    y = target
    plt.scatter(x, y, marker="o")
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Hours Studied")
```
